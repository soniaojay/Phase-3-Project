**Handling Outliers**

Use of **IQR method** to remove outliers for variables where outliers are clearly visible and can impact the analysis.
As well as **Log Transformation** has been used to reduce the skewness for a higly skewed data: number vmail messages, 'total intl calls, 'customer service calls. It is particularly useful for right-skewed distributions. However, note that log transformation can only be applied to positive values.

```python
def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Filter out the outliers
    df_filtered = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    return df_filtered

# List of columns to remove outliers
columns_with_outliers = ['account length','number vmail messages', 'total day minutes', 'total day calls', 'total day charge', 'total eve minutes', 'total eve calls', 'total eve charge', 'total night minutes', 'total night calls', 'total night charge', 'total intl minutes', 'total intl calls', 'total intl charge', 'customer service calls', 'total_minutes_within', 'total_minutes_overall', 'total_calls_within', 'total_calls_overall', 'total_charge_within', 'total_charge_overall', 'charge_per_call_intl', 'charge_per_call_within', 'charge_per_call_night', 'charge_per_call_eve', 'charge_per_call_day', 'customer_call_satisfaction']
Tel_df_copy_filtered = Tel_df_copy.copy()  # Creating a copy to avoid overwriting the original data

# Remove outliers
for column in columns_with_outliers:
    Tel_df_copy_filtered = remove_outliers(Tel_df_copy_filtered, column)

# List of columns to apply log transformation as they are positively skewed
columns_to_log_transform = ['charge_per_call_intl','number vmail messages', 'total intl calls', 'customer service calls','charge_per_call_night', 'charge_per_call_eve', 'charge_per_call_day', 'customer_call_satisfaction']

# Apply log transformation
for column in columns_to_log_transform:
    Tel_df_copy_filtered[column] = np.log1p(Tel_df_copy_filtered[column])  

# Check the dimensions of the dataframe after outlier removal and log transformation
print(Tel_df_copy_filtered.shape)
```

### 4. Model building on the Telco Churn Dataset

```python
#Import dependencies
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split, StratifiedKFold, cross_val_score, cross_val_predict
from sklearn.metrics import precision_score, accuracy_score, recall_score, f1_score
from sklearn.linear_model import LogisticRegression
from sklearn.compose import ColumnTransformer
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
from imblearn.over_sampling import SMOTE
from imblearn.pipeline import Pipeline
```

 ### 4.1 Logistic Regression

**Create a Pipeline with SMOTE and Logistic Regression**

```python
X = Tel_df_copy_filtered[['customer_call_satisfaction','international plan','total day minutes', 'total day charge', 'total eve minutes', 'total eve charge','total intl minutes', 'total intl charge','total_minutes_within', 'total_minutes_overall', 'total_charge_within', 'total_charge_overall', 'charge_per_call_intl', 'charge_per_call_within', 'charge_per_call_eve', 'charge_per_call_day']]
y = Tel_df_copy_filtered['churn']
print(X.shape)
print(y.shape)

#Feature Scaling through Standardization
scaler = StandardScaler().fit(X)
scaled_X = scaler.transform(X)

# Define the pipeline
pipeline = Pipeline([
    ('smote', SMOTE(random_state=42)),
    ('log_reg', LogisticRegression())
])
```

**Evaluate the Model using Cross-Validation**

```python
# Define the cross-validation strategy
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# Perform cross-validation and get predicted probabilities
y_pred_prob_cv = cross_val_predict(pipeline, scaled_X, y, cv=cv, method='predict_proba')[:, 1]

# Convert probabilities to binary predictions
y_pred_cv = (y_pred_prob_cv > 0.5).astype(int)

# Evaluate the model using cross-validation predictions
accuracy_cv = accuracy_score(y, y_pred_cv)
conf_matrix_cv = confusion_matrix(y, y_pred_cv)
class_report_cv = classification_report(y, y_pred_cv)

print("Cross-validation accuracy:", accuracy_cv)
print("--------------------------------------")
print("Cross-validation Confusion Matrix:\n", conf_matrix_cv)
print("--------------------------------------")
print("Cross-validation Classification Report:\n", class_report_cv)

# Create a heatmap for the confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix_cv, annot=True, fmt='d', cmap='Blues', xticklabels=['Not Churn', 'Churn'], yticklabels=['Not Churn', 'Churn'])
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.title('Confusion Matrix - Cross-Validation')
plt.show()
```

**Interpreting the results obtained through cross-validation**

**1. Accuracy:**
An accuracy of 80.4% suggests that the Logistic Regression model correctly predicts the class labels for around 80.5% of the instances in the dataset.

**Precision:** The model achieves a precision of 97% for the negative class (0) and 31% for the positive class (1). This suggests that when the model predicts non-churn (0), it is correct about 97% of the time, and when it predicts churn (1), it is correct about 31% of the time.

**Recall:** The model achieves a recall of 81% for the negative class (0) and 78% for the positive class (1). This indicates that the model is effective at capturing a substantial portion of the positive cases.

**F1-Score:** A higher F1 Score indicates better overall performance in terms of precision and recall. The model achieves an F1 Score of 88% for the negative class (0) and 44% for the positive class (1).

**Conclusion:**
The model demonstrates reasonably good performance in predicting churn, as indicated by the relatively high recall and balanced F1-score. However, there is room for improvement, especially in reducing false positives and enhancing precision for churn.

### Adopt Low Fixed Charge

We know that total charge is one of the features of the dataset that has has high influence in our model during EDA.

```python
#see the statistics for total charge and minutes used overall (within the country and intl) calls for those who churn
Tel_df_copy_filtered[Tel_df_copy_filtered['churn'] == 1][['total_charge_overall','total_minutes_overall']].describe()
```

At 50% percentile of total_minutes_overall is ~700 minutes for customers who churn. We also see that the overall charge at this percentile is ~$71.05.  

Implementation of a fixed charge for those who go over 700 minutes at a price of $71.05. The logistic regression model was selected to be used here as the model had the best predictor for the training and testing data for the recall. The recall score was close unlike the decision tree and random forest where the training set performed far better than the test data.

We created the function below to allow us to test different overall minute and overall charge parameters and how it affects churn.

```python

def predict_churn(total_minutes_cap, charge_per_minute):
    # Cap the 'total_minutes_overall' at the specified percentile
    Tel_df_copy_filtered['total_minutes_overall_capped'] = np.minimum(Tel_df_copy_filtered['total_minutes_overall'], total_minutes_cap)
    
    # Prepare the feature set and target variable
    X = Tel_df_copy_filtered[['international plan', 'total day minutes', 'total day charge', 'total eve minutes', 'total eve charge',
                              'total intl minutes', 'total intl charge', 'total_minutes_within', 'total_minutes_overall_capped', 
                              'total_charge_within', 'total_charge_overall', 'charge_per_call_intl', 'charge_per_call_within', 
                              'charge_per_call_eve', 'charge_per_call_day', 'customer_call_satisfaction']]
    y = Tel_df_copy_filtered['churn']

    # Feature Scaling through Standardization
    scaler = StandardScaler().fit(X)
    scaled_X = scaler.transform(X)

    # Define the pipeline
    pipeline = Pipeline([
        ('smote', SMOTE(random_state=42)),
        ('log_reg', LogisticRegression(random_state=42, max_iter=1000))
    ])

    # Define the cross-validation strategy
    cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

    # Perform cross-validation and get predicted probabilities
    y_pred_cv = cross_val_predict(pipeline, scaled_X, y, cv=cv)

    # Fit the pipeline to the entire dataset to extract coefficients
    pipeline.fit(scaled_X, y)

    # Predict the churn values on the entire dataset
    churn_prediction = pipeline.predict(scaler.transform([[0, 0, 0, 0, 0, 0, 0, 0, total_minutes_cap, 0, charge_per_minute, 0, 0, 0, 0, 0]]))[0]
    return churn_prediction

# Example usage
total_minutes_cap = 700
charge_per_minute = 71.05
churn_prediction = predict_churn(total_minutes_cap, charge_per_minute)
print("Predicted churn value:", churn_prediction)
```

**After applying our fixed fee plan, churn decreased from 14.49% to 0.** 

The predicted churn value of 0 indicates that, according to the model, the customer is not likely to churn. In other words, based on the features provided (including capped total minutes and charge per minute), the model suggests that the customer is expected to remain with the telecom service. This prediction could be valuable for decision-making processes related to customer retention strategies and resource allocation within the telecom company.
